```python
%%R
library(VennDiagram)
library("gridExtra")
library("extrafont")
font_import()


count_version_1 =  9883
count_version_2 = 10379
count_version_3 = 2680
count_version_1_and_2 = 145
count_version_1_and_3 =  91
count_version_2_and_3 =  142
count_version_1_and_2_and_3 =  20

set1 <- count_version_1 + count_version_1_and_2 + count_version_1_and_3 + count_version_1_and_2_and_3
set2 <- count_version_2 + count_version_1_and_2 + count_version_2_and_3 + count_version_1_and_2_and_3
set3 <- count_version_3 + count_version_1_and_3 + count_version_2_and_3 + count_version_1_and_2_and_3
```

```python
%%R
loadfonts()
fontfamily = "Arial"
main.fontfamily="serif"
my_venndiag <- draw.triple.venn(area1 = set1, area2 = set2, area3 = set3, n12 = count_version_1_and_2, n23 = count_version_2_and_3, n13 = count_version_1_and_3, 
    n123 = count_version_1_and_2_and_3, category = c("Version 1", "Version 2", "Version 3" ), lty = "blank", 
    fill = c("purple", "orangered", "darkgoldenrod"), fontfamily = "Arial", main.fontfamily="serif")
```

```python
%%R
library(grDevices)
pdf(file="venn.pdf")
grid.arrange(gTree(children = my_venndiag), # Add title & subtitle
             top = "Number of ROR Providers Who Have Completed Trainings")
```
