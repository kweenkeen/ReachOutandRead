```python
import pandas as pd
from google.colab import files
from datetime import date
from datetime import datetime

# UPLOAD SQC DATA REPORT

SQC_survey_report = files.upload()
SQC_survey_file = next(iter(SQC_survey_report))
df = pd.read_excel(SQC_survey_file, converters={'Site ID': str})
# df = df.set_index('Site ID')
```

```python
today = str(date.today())
fiscal_year = input("Type in desired Fiscal Year (YY): " )
filename = 'SQC_FY' + fiscal_year + '_' + today + '.xlsx'
writer = pd.ExcelWriter(filename, engine='xlsxwriter')
```

```python
date_range_min = input("Type in the start of your year (MM/DD/YYYY): ")
date_range_min = datetime.strptime(date_range_min, '%m/%d/%Y')
date_range_max = input("Type in the end of your year (MM/DD/YYYY): ")
date_range_max = datetime.strptime(date_range_max, '%m/%d/%Y')
```

```python
assessment_date_df = df.loc[(df['Assessment Date'] >= date_range_min) & (df['Assessment Date'] <= date_range_max)]
assessment_date_df['Assessment Date'] = assessment_date_df['Assessment Date'].dt.date
# classification_df = assessment_date_df['Classification'].value_counts()
```

```python
oregon_df = assessment_date_df[assessment_date_df['Affiliate Name'] == 'Reach Out and Read Oregon']
oregon_df.to_excel(writer, sheet_name = "Oregon Dataframe")
oregon_classification_df = oregon_df['Classification'].value_counts()
oregon_classification_df = pd.DataFrame({'Classification':oregon_classification_df.index, '# of Sites':oregon_classification_df.values})
oregon_classification_df.to_excel(writer, sheet_name = "Oregon Summary")
```
