```python
!pip install xlsxwriter
import pandas as pd
from datetime import date
import zipfile
from google.colab import files
import matplotlib.pyplot as plt
```

```python
progress_report = files.upload()
progress_report = next(iter(progress_report))
progress_df = pd.read_excel(progress_report, converters={'Site ID': str})
```

```python
## UPLOAD SITE DATA REPORT

site_data_report = files.upload()
site_data_report = next(iter(site_data_report))
site_data_df = pd.read_excel(site_data_report, converters={'Site ID': str})
```

```python
# CHANGE INDEX TO SITE ID
progress_df = progress_df.set_index('Site ID')
site_data_df = site_data_df.set_index('Site ID')
```

```python
# APPLY FILTERS

# Filter for high level active sites
high_level_active = ['Active', 'Active - Non-Reporting', 'Slated for deactivation', 'COVID-19 Freeze'] 
progress_df = progress_df[progress_df['Status'].isin(high_level_active)] 
site_data_df = site_data_df[site_data_df['Status'].isin(high_level_active)]
```

```python
# CREATE LIST OF INSURANCE COLUMN NAMES

insurance_cols = [col for col in progress_df.columns if 'Insurance' in col]
race_ethnicity_cols = [col for col in progress_df.columns if 'Race' in col]
language_cols = [col for col in progress_df.columns if 'Language' in col]
```

```python
demographic_cols = insurance_cols + race_ethnicity_cols + language_cols
```

```python
site_data_df.head()
```

```python
site_data_df = site_data_df.reindex(columns = site_data_df.columns.tolist() + demographic_cols)
```

```python
site_data_df.head()
```

```python
for i in demographic_cols:
  for j, rows in site_data_df.iterrows():
    site_data_df.loc[j, i] = progress_df.loc[j, i]
```

```python
site_data_df.head()
```

```python
# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('Site_Data_Impact.xlsx', engine='xlsxwriter')
```

```python
## Filter out for Active Sites
site_data_active_df = site_data_df.loc[(site_data_df['Status'] == 'Active')]
site_data_active_df = site_data_active_df['Affiliate Name'].value_counts().rename_axis('Affiliate').reset_index(name='Active Sites')
site_data_active_df.head()
```

```python
site_data_active_nonreporting_df = site_data_df.loc[(site_data_df['Status'] == 'Active - Non-Reporting')]
site_data_active_nonreporting_df = site_data_active_nonreporting_df['Affiliate Name'].value_counts().rename_axis('Affiliate').reset_index(name='Active - Non-Reporting Sites')
site_data_active_nonreporting_col = site_data_active_nonreporting_df['Active - Non-Reporting Sites']
site_data_active_df = site_data_active_df.join(site_data_active_nonreporting_col)
site_data_active_df.head()
```

```python
site_data_CovidFreeze_df = site_data_df.loc[(site_data_df['Status'] == 'COVID-19 Freeze')]
site_data_CovidFreeze_df = site_data_CovidFreeze_df['Affiliate Name'].value_counts().rename_axis('Affiliate').reset_index(name='COVID-19 Freeze Sites')
site_data_CovidFreeze_col = site_data_CovidFreeze_df['COVID-19 Freeze Sites']
site_data_active_df = site_data_active_df.join(site_data_CovidFreeze_col)
site_data_active_df.head()
```

```python
site_data_S4D_df = site_data_df.loc[(site_data_df['Status'] == 'Slated for deactivation')]
site_data_S4D_df = site_data_S4D_df['Affiliate Name'].value_counts().rename_axis('Affiliate').reset_index(name='Slated for Deactivation Sites')
site_data_S4D_col = site_data_S4D_df['Slated for Deactivation Sites']
site_data_active_df = site_data_active_df.join(site_data_S4D_col)
site_data_active_df.head()
```

```python
site_data_total_active_df = site_data_df['Affiliate Name'].value_counts().rename_axis('Affiliate').reset_index(name='Total High Level Active Sites')
site_data_total_active_col = site_data_total_active_df['Total High Level Active Sites']
site_data_active_df = site_data_active_df.join(site_data_total_active_col)
site_data_active_df.head()
```

```python
site_data_primary_df = site_data_df.loc[(site_data_df['Site Type'] == 'Primary')]
site_data_primary_df = site_data_primary_df['Affiliate Name'].value_counts().rename_axis('Affiliate').reset_index(name='Number of Primary Sites')
site_data_primary_df['Number of Primary Sites'] = site_data_primary_df['Number of Primary Sites'] / site_data_total_active_col
site_data_primary_col = site_data_primary_df['Number of Primary Sites']
site_data_active_df = site_data_active_df.join(site_data_primary_col)
site_data_active_df.head()
```

```python
site_data_satellite_df = site_data_df.loc[(site_data_df['Site Type'] == 'Satellite')]
site_data_satellite_df = site_data_satellite_df['Affiliate Name'].value_counts().rename_axis('Affiliate').reset_index(name='Number of Satellite Sites')
site_data_satellite_df['Number of Satellite Sites'] = site_data_satellite_df['Number of Satellite Sites'] / site_data_total_active_col
site_data_satellite_col = site_data_satellite_df['Number of Satellite Sites']
site_data_active_df = site_data_active_df.join(site_data_satellite_col)
site_data_active_df.head()
```
