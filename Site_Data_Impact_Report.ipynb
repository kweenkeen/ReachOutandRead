```python
!pip install xlsxwriter
import pandas as pd
from datetime import date
import zipfile
from google.colab import files
import matplotlib.pyplot as plt
```

```python
progress_report = files.upload()
progress_report = next(iter(progress_report))
progress_df = pd.read_excel(progress_report, converters={'Site ID': str})
```

```python
## UPLOAD SITE DATA REPORT FOR *TWO* REPORTING PERIODS

site_data_report = files.upload()
site_data_report = next(iter(site_data_report))
site_data_df = pd.read_excel(site_data_report, converters={'Site ID': str})
```

```python
today = str(date.today())
filename = 'Site_Impact_' + today + '.xlsx'
writer = pd.ExcelWriter(filename, engine='xlsxwriter')
```

```python
# CHANGE INDEX TO SITE ID
progress_df = progress_df.set_index('Site ID')
site_data_df = site_data_df.set_index('Site ID')
```

```python
# APPLY FILTERS

# Filter for high level active sites
high_level_active = ['Active', 'Active - Non-Reporting', 'Slated for deactivation', 'COVID-19 Freeze'] 
progress_df = progress_df[progress_df['Status'].isin(high_level_active)] 
site_data_df = site_data_df[site_data_df['Status'].isin(high_level_active)]
```

```python
# CREATE LIST OF INSURANCE COLUMN NAMES

insurance_cols = [col for col in progress_df.columns if 'Insurance' in col]
race_ethnicity_cols = [col for col in progress_df.columns if 'Race' in col]
language_cols = [col for col in progress_df.columns if 'Language' in col]
```

```python
demographic_cols = insurance_cols + race_ethnicity_cols + language_cols
```

```python
site_data_df.head()
```

```python
site_data_df = site_data_df.reindex(columns = site_data_df.columns.tolist() + demographic_cols)
```

```python
site_data_df.head()
```

```python
for i in demographic_cols:
  for j, rows in site_data_df.iterrows():
    site_data_df.loc[j, i] = progress_df.loc[j, i]
```

```python
site_data_df.head()
```

```python
site_data_impact_df = pd.DataFrame(columns=['Affiliate Name'])
site_data_impact_df = site_data_impact_df.set_index('Affiliate Name')
site_data_impact_df.head()
```

```python
## Filter out for Active Sites
site_data_active_df = site_data_df.loc[(site_data_df['Status'] == 'Active')]
site_data_active_df = site_data_active_df['Affiliate Name'].value_counts()
site_data_active_df = site_data_active_df.sort_index()
site_data_impact_df['Active Sites'] = site_data_active_df
site_data_impact_df.head()
```

```python
site_data_active_nonreporting_df = site_data_df[site_data_df['Status'] == 'Active - Non-Reporting']
site_data_active_nonreporting_df = site_data_active_nonreporting_df['Affiliate Name'].value_counts()
site_data_active_nonreporting_df = pd.DataFrame({'Affiliate Name':site_data_active_nonreporting_df.index, 'Active - Non-Reporting':site_data_active_nonreporting_df.values})
site_data_impact_df = site_data_impact_df.join(site_data_active_nonreporting_df.set_index('Affiliate Name'), on='Affiliate Name')
site_data_impact_df['Active - Non-Reporting'] = site_data_impact_df['Active - Non-Reporting'].fillna(0)
site_data_impact_df['Active - Non-Reporting'] = site_data_impact_df['Active - Non-Reporting'].astype(int)
site_data_impact_df.head()
# site_data_active_nonreporting_df.head()
```

```python
site_data_covid_freeze_df = site_data_df[site_data_df['Status'] == 'COVID-19 Freeze']
site_data_covid_freeze_df = site_data_covid_freeze_df['Affiliate Name'].value_counts()
site_data_covid_freeze_df = pd.DataFrame({'Affiliate Name':site_data_covid_freeze_df.index, 'COVID Freeze':site_data_covid_freeze_df.values})
site_data_impact_df = site_data_impact_df.join(site_data_covid_freeze_df.set_index('Affiliate Name'), on='Affiliate Name')
site_data_impact_df['COVID Freeze'] = site_data_impact_df['COVID Freeze'].fillna(0)
site_data_impact_df['COVID Freeze'] = site_data_impact_df['COVID Freeze'].astype(int)
site_data_impact_df.head()
```

```python
site_data_S4D_df = site_data_df[site_data_df['Status'] == 'Slated for deactivation']
site_data_S4D_df = site_data_S4D_df['Affiliate Name'].value_counts()
site_data_S4D_df = pd.DataFrame({'Affiliate Name':site_data_S4D_df.index, 'Slated for Deactivation':site_data_S4D_df.values})
site_data_impact_df = site_data_impact_df.join(site_data_S4D_df.set_index('Affiliate Name'), on='Affiliate Name')
site_data_impact_df['Slated for Deactivation'] = site_data_impact_df['Slated for Deactivation'].fillna(0)
site_data_impact_df['Slated for Deactivation'] = site_data_impact_df['Slated for Deactivation'].astype(int)
site_data_impact_df.head()
```

```python
high_level_active = ['Active', 'Active - Non-Reporting', 
                     'Slated for deactivation', 'COVID-19 Freeze'] 
site_data_total_active_df = site_data_df[site_data_df['Status'].isin(high_level_active)]
site_data_total_active_df = site_data_total_active_df['Affiliate Name'].value_counts()
site_data_total_active_df = pd.DataFrame({'Affiliate Name':site_data_total_active_df.index, 'Total High Level Active Sites':site_data_total_active_df.values})
site_data_impact_df = site_data_impact_df.join(site_data_total_active_df.set_index('Affiliate Name'), on='Affiliate Name')
site_data_impact_df['Total High Level Active Sites'] = site_data_impact_df['Total High Level Active Sites'].fillna(0)
site_data_impact_df['Total High Level Active Sites'] = site_data_impact_df['Total High Level Active Sites'].astype(int)
site_data_impact_df.head()

```

```python
site_data_primary_df = site_data_df.loc[(site_data_df['Site Type'] == 'Primary')]
site_data_primary_df = site_data_primary_df['Affiliate Name'].value_counts()
site_data_primary_df = pd.DataFrame({'Affiliate Name':site_data_primary_df.index, 'Number of Primary Sites':site_data_primary_df.values})
site_data_primary_df.head()
site_data_impact_df = site_data_impact_df.join(site_data_primary_df.set_index('Affiliate Name'), on='Affiliate Name')
site_data_impact_df['Number of Primary Sites'] = site_data_impact_df['Number of Primary Sites'].fillna(0)
site_data_impact_df['Number of Primary Sites'] = site_data_impact_df['Number of Primary Sites'].astype(int)
site_data_impact_df.head()
```

```python
site_data_satellite_df = site_data_df.loc[(site_data_df['Site Type'] == 'Satellite')]
site_data_satellite_df = site_data_satellite_df['Affiliate Name'].value_counts()
site_data_satellite_df = pd.DataFrame({'Affiliate Name':site_data_satellite_df.index, 'Number of Satellite Sites':site_data_satellite_df.values})
site_data_satellite_df.head()
site_data_impact_df = site_data_impact_df.join(site_data_satellite_df.set_index('Affiliate Name'), on='Affiliate Name')
site_data_impact_df['Number of Satellite Sites'] = site_data_impact_df['Number of Satellite Sites'].fillna(0)
site_data_impact_df['Number of Satellite Sites'] = site_data_impact_df['Number of Satellite Sites'].astype(int)
site_data_impact_df.head()
```

```python
site_data_impact_df['% Primary Sites'] = site_data_impact_df['Number of Primary Sites'] / site_data_impact_df['Total High Level Active Sites']
site_data_impact_df['% Primary Sites'] = site_data_impact_df['% Primary Sites']* 100
site_data_impact_df['% Primary Sites'] = site_data_impact_df['% Primary Sites'].round(2)
site_data_impact_df['% Primary Sites'] = site_data_impact_df['% Primary Sites'].astype(str) + '%'
site_data_impact_df['% Satellite Sites'] = site_data_impact_df['Number of Satellite Sites'] / site_data_impact_df['Total High Level Active Sites']
site_data_impact_df['% Satellite Sites'] = site_data_impact_df['% Satellite Sites']* 100
site_data_impact_df['% Satellite Sites'] = site_data_impact_df['% Satellite Sites'].round(2)
site_data_impact_df['% Satellite Sites'] = site_data_impact_df['% Satellite Sites'].astype(str) + '%'
site_data_impact_df.head()
```

```python
site_data_impact_df.to_excel(writer, sheet_name = "Site Impact FY21 Impact") 
```

```python
writer.save()
files.download(filename)
```
